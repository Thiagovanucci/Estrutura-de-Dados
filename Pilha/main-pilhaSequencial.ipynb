```python

class Pilha:
    def __init__(self,tamanho):
        self.tammax = tamanho
        self.dados=[]
        self.topo = -1
    def pilhaVazia(self):
        if self.topo == -1:
            print("está vazia\n")

        else:
            print("não está vazia\n")
    def pilhaCheia(self):
        if self.topo == self.tammax -1:
            print("A pilha está cheia")
        else:
            print("A pilha nao está cheia")
    def tamanhoPilha(self):
        return self.topo +1
    def push(self,conteudo):
        if self.topo == self.tammax-1 :
            return print("A pilha está cheia")
        self.dados.append(conteudo)
        self.topo+=1
    def pop(self):
        if self.topo == -1 :
            return print("a pilha está vazia")
        del self.dados[self.topo]
        self.topo-=1
    def top(self):
        if self.topo == -1 :
            return print("a pilha está vazia")
        return self.dados[self.topo]
        

        
        
        
teste =Pilha(5)
teste.pilhaVazia()
teste.pilhaCheia()
teste.push(10)
print("inserindo 10")

teste.push(65)
print("inserindo 65")

teste.push(32)
print("inserindo 32")
print(f"tamnaho da pilha: {teste.tamanhoPilha()}")
print(f"topo da pilha: {teste.top()}")

teste.push(39)
print("inserindo 39")
teste.push(16)
print("inserindo 16")
print(f"tamnaho da pilha: {teste.tamanhoPilha()}")
print(f"topo da pilha: {teste.top()}")

print("inserindo 82")
teste.push(82)

teste.pop()
print("movendo da pilha")
print(f"tamnaho da pilha: {teste.tamanhoPilha()}")
print(f"topo da pilha: {teste.top()}")
```
